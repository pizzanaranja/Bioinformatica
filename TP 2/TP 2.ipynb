```python
#Ejercicio 2.2

import Bio.SearchIO as bpio

filas = []
resultados = list(bpio.parse("n315_vfdb.xml","blast-xml"))
for result_query in resultados:
    query_len = result_query.seq_len
    for hit in result_query:
        hit_len = hit.seq_len
        for hsp in hit.hsps:
            qcov = hsp.aln_span / query_len * 100
            hcov = hsp.aln_span / hit_len * 100
            identidad = hsp.ident_num / hsp.aln_span * 100
            descripcion = hit.description
            especie = descripcion.split("OS=")[-1].split(" ")[0] if "OS=" in descripcion else "NA"
            filas.append({"query": result_query.id,
                          "hit": hit.id,
                          "cov query": qcov,
                          "cov hit": hcov,
                          "identidad": identidad,
                          "especie": especie})
df = pd.DataFrame(filas)
df.head()
```
