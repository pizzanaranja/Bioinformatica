I) En la sección "alignments" se puede ver que ambas comparaciones toman los 154 aminoácidos, lo que significa que es un alineamiento global.

II) No hay gaps en los dot-plot, solo matches y mismatches. Si hubieran gaps, se verían como huecos entre lineas continuas.

III) Comparando con la bovina, el alineamiento óptimo tiene un score de 267, que es muy alto y un E-value de 2e-98, que es extremadamente bajo. En identities se ve que matchean el 84% de aminoácidos, y el 89% son aminoácidos similares. Hay 0 gaps.
Comparando con la porcina, se ven resultados similares al caso anterior, un score de 294, un E-value aún mas bajo (5e-109), 94% de matches, 96% de aminoácidos similares y 0 gaps.

Si cambio a BLOSUM80, los alineamientos se mantienen igual, pero en ambos casos el score aumenta, el E-value disminuye, y las identidades y positives se mantienen casi igual (varían 1%).

Si cambio los valores de gap existence y extension a 11:2, sucede lo mismo que antes, pero el E-value disminuye todavía más.

Si cambio los valores de gap existence y extension a 9:1, pasa lo opuesto. Los E-values aumentan bastante para ambos casos, y el score disminuye.

Cuando cambiamos a BLOSUM80 los resultados observados concuerdan con el hecho de que son proteínas con mucha similitud, y BLOSUM80 sirve para esas comparaciones, aumentando los score por match y penalizando los gaps (que en estos casos no existieron). Luego, cuando usamos 11:2, estamos haciendo que la evaluación sea mas estricta, y como siguen sin haber gaps, el score general aumenta. Finalmente cuando usamos 9:1, estamos siendo menos estrictos con la evaluación, por lo que disminuye el score.

En cambio al usar la base refseq_select directamente no hay repeticiones (isoformes).

II) Citoglobina: Se cree que promueve la difusión de oxígeno en los tejidos. El alineamiento con la mioglobina da un score de 73.6 en BLAST, y un E-value de 5e-17, lo que nos indica que es muy probable que haya una relación biológica entre ambas proteínas.

Hemoglobina subunidad z: Proteína esencial para el transporte de oxígeno en las primeras semanas de gestación. El alineamiento con la mioglobina tiene un score de 71.6 en BLAST, y con un E-value de 1e-16 se puede afirmar que comparten relación biológica.

III) Según BLAST, la proteína mas parecida es la citoglobina, ya que es la de mayor score, menor E-value y mayores valores de identidad. No es sorprendente ya que es una globina también.

IV) No creemos que estas sean todas las globinas, ya que BLAST puede haber omitido algunas por no ser del todo similares. Pueden haber globinas que hayan evolucionado para tener una función distinta a la mioglobina.
